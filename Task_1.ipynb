Prediction using Supervised ML

Importing Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

Importing Dataset

```python
# Reading data from remote link
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
data.head(10)
```

```python
data.describe()
```

Checking Null Values


```python
data.info()
```

Visualizing

```python
plt.scatter(data['Hours'], data['Scores'])
plt.title('Hours VS Scores Scatter Plot', size=16)
plt.xlabel('Hours of Study', size=16)
plt.ylabel('Scores', size=16)
plt.show()
```

```python
x=data.iloc[:, :-1].values
y=data.iloc[:, -1].values
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.25)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
lr= LinearRegression()
```

```python
model=lr.fit(x_train, y_train)
```

```python
plt.scatter(x_train, y_train)
plt.title('Training Set: Hours Vs Scores', size=16)
plt.xlabel('Hours', size=16)
plt.ylabel('Scores', size=16)
plt.show()
```

```python
pred= model.predict(x_test)
```

```python
pred_y= pd.DataFrame({'Actual':y_test, 'Predicted': pred, 'Difference': y_test-pred})
pred_y
```

```python
model.score(x_test, y_test)
```

Model has 95% Accuracy

```python
pred0= model.predict([[9.25]])
print('The predicted score if the student studies for 9.25 hrs/day is', pred0)
```

From this,if the student studies 9.25 hrs/day it can score 92.07%
